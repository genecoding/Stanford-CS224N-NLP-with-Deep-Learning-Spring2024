As a reminder, dot product attention is $e_{t,i} = s^{T}_{t}h_{i}$, multiplicative attention is $e_{t,i} = s^{T}_{t} Wh_{i}$, and additive attention is $e_{t,i} = v^{T}tanh(W_{1}h_{i} + W_{2}s_{t})$.
* i. Explain one advantage and one disadvantage of *dot product attention* compared to multiplicative attention.  
  (adv) Dot product attention is simpler and faster.  
  (dis) Dot product attention is less flexible. By adding an extra learnable matrix, multiplicative attention can learn to better capture complex interactions between query and values.
* ii. Explain one advantage and one disadvantage of *additive attention* compared to multiplicative attention.  
  (adv) By adding more learnable metrices and a non-linear activation function, additive attention can thus handle more complex, non-linear relationships between query and values.  
  (dis) Additive attention is more computationally expensive and harder to optimize with efficient matrix multiplication compared to multiplicative attention.
